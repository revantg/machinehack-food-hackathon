```python
data['COST_INTERVAL'] = pd.cut(data['COST'], 50)
```

```python
titles_data = data.loc[data["TITLE"].notnull(), ["TITLE", "COST_INTERVAL"]]
```

```python
grouped_titles_data = titles_data.groupby(by = ['COST_INTERVAL'])
```

```python
most_used_titles = {}
for i in grouped_titles_data.groups.keys():
    temp_title_counts = []
    try :
        grouped_titles_data.get_group(i)['TITLE'].apply(lambda x: [temp_title_counts.append(i) for i in x.split(",")])
        most_used_titles[i] = pd.Series(temp_title_counts).value_counts().index[0]
    except:
        print("Failed for", i)
        pass
#     break
```

```python
data.loc[data['TITLE'].isnull(), 'TITLE'] = data.loc[data['TITLE'].isnull()]['COST_INTERVAL'].apply(lambda x : most_used_titles[x] if x in most_used_titles else np.nan)
```

```python
non_null_titles_data = data[data['TITLE'].notnull()]
grouped_non_null_titles = non_null_titles_data[['TITLE', 'CUISINES', 'COST_INTERVAL']].groupby(by = ["COST_INTERVAL"])
```

```python
titles_to_fill = []
for cost_interval, cuisines in data.loc[data['TITLE'].isnull(), ["COST_INTERVAL", "CUISINES"]].values:
    cuisines = cuisines.split(",")

    df = grouped_non_null_titles.get_group(cost_interval)
    
    mx_dict = {k : [] for k in range(10)}
    
    for i, j in df[['CUISINES', 'TITLE']].values:
        tmp_cuisines = i.split(",")
        mx_dict[len(np.intersect1d(tmp_cuisines, cuisines))].append(j)
    
    mx_occuring_title = np.nan
    for i in range(9, -1, -1):
        if len(mx_dict[i]):
            mx_occuring_title = pd.Series(mx_dict[i]).value_counts().index[0]
            break
    
    titles_to_fill.append(mx_occuring_title)
        
    
#     print(mx, element)
#     break
```

```python
for i in range(4, -1, -1):
    if len(mx_dict[i]):
        print(pd.Series(mx_dict[i]).value_counts().index[0])
        break
```

```python
data.loc[data['TITLE'].isnull(),"TITLE"] = titles_to_fill
```
